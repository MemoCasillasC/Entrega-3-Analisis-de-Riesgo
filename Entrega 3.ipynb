4. Comparación con el índice más adecuado (por ejemplo: SP500, MSCI World, IPC) usando métricas: rendimiento acumulado, drawdown máximo, volatilidad

**Para el activo: VISA (V)**

```python
data = yf.download(["V", "^GSPC"], start=start_date, end = end_date, auto_adjust=True)["Close"].dropna()
px_v = data["V"]
px_sp = data["^GSPC"]

ret_v = np.log(px_v).diff().dropna()
ret_sp = np.log(px_sp).diff().dropna()

# Rendimiento acumulado
rend_v = np.exp(ret_v.cumsum())[-1] - 1
rend_sp = np.exp(ret_sp.cumsum())[-1] - 1

# Volatilidad anualizada
vol_v = ret_v.std() * np.sqrt(252)
vol_sp = ret_sp.std() * np.sqrt(252)

# Drawdown máximo
def max_dd(serie):
    cummax = serie.cummax()
    dd = serie / cummax - 1
    return dd.min()

wealth_v = np.exp(ret_v.cumsum())
wealth_sp = np.exp(ret_sp.cumsum())
dd_v = max_dd(wealth_v)
dd_sp = max_dd(wealth_sp)

# Resultados
print("=== VISA (V) ===")
print(f"Rendimiento acumulado: {rend_v:.2%}")
print(f"Volatilidad anualizada: {vol_v:.2%}")
print(f"Drawdown máximo: {dd_v:.2%}\n")

print("=== S&P500 ===")
print(f"Rendimiento acumulado: {rend_sp:.2%}")
print(f"Volatilidad anualizada: {vol_sp:.2%}")
print(f"Drawdown máximo: {dd_sp:.2%}")
```

```python
wealth_v.plot(label="Visa", color = "darkviolet")
wealth_sp.plot(label="S&P500", color = "palevioletred")
plt.title("Rendimiento acumulado")
plt.legend()
```

## Entrega 3

**Para el activo: VISA (V)**

    Estimación de rendimientos (porcentuales) y volatilidad (estimadores: sample std, estimador anualizado)

```python
# --- Parámetros ---
periods_per_year = 252  # asumiendo datos diarios

# --- Diccionario con tus series ---
# Ajusta con tus nombres de variables reales:
assets = {
    'VISA': d_V['Close']['V'],
    'SAMSUNG': d_S['Close']['005930.KS'],
    'MC DONALDS': d_M['Close']['MCD'],
    'Eli Lilly': d_E['Close']['LLY'],
    'COSTCO': d_C['Close']['COST']
}

# --- Lista para acumular resultados ---
rows = []

# --- Loop sobre cada activo ---
for name, prices in assets.items():
    # Calcular rendimientos porcentuales
    returns = prices.pct_change().dropna()

    # Promedio y volatilidad (estimador muestral)
    mean_daily = returns.mean()
    std_daily = returns.std(ddof=1)

    # Annualización
    mean_annual = mean_daily * periods_per_year
    std_annual = std_daily * np.sqrt(periods_per_year)

    # Guardar resultados en lista de diccionarios
    rows.append({
        'Activo': name,
        'Mean_Daily': mean_daily,
        'Std_Daily': std_daily,
        'Mean_Annual': mean_annual,
        'Std_Annual': std_annual
    })

# --- Convertir a DataFrame final ---
summary_df = pd.DataFrame(rows)
summary_df.set_index('Activo', inplace=True)

port_row = pd.DataFrame({
    'Mean_Daily': [media_retorno_diario],
    'Std_Daily': [vol_diaria],
    'Mean_Annual': [media_retorno_anual],
    'Std_Annual': [volatilidad_anual]
}, index=['PORTFOLIO'])

# Agregarla al DataFrame existente
summary_df = pd.concat([summary_df, port_row])


# --- Mostrar resultados ---

summary_df.round(6)
```

    Pruebas de normalidad sobre: rendimientos, precios, y log(precios). Usar pruebas estadísticas (Jarque-Bera, Kolmogorov–Smirnov, Shapiro–Wilk cuando aplique) y QQ-plots.

```python
simport statsmodels.api as sm
from scipy import stats
from statsmodels.stats.stattools import jarque_bera

# ---------- Funciones de pruebas ----------
def ks_test_standardized(x: pd.Series):
    x = x.dropna()
    if len(x) < 10:
        return np.nan, np.nan
    z = (x - x.mean()) / x.std(ddof=1)
    stat, pval = stats.kstest(z, "norm")
    return stat, pval

def shapiro_safe(x: pd.Series, max_n=5000):
    x = x.dropna()
    if len(x) < 3:
        return np.nan, np.nan
    if len(x) > max_n:
        x = x.sample(max_n, random_state=42)
    stat, pval = stats.shapiro(x.values)
    return stat, pval

def jarque_bera_test(x: pd.Series):
    x = x.dropna()
    if len(x) < 10:
        return np.nan, np.nan, np.nan, np.nan
    jb_stat, jb_pvalue, skew, kurt = jarque_bera(x.values)
    return jb_stat, jb_pvalue, skew, kurt
    
def qqplot(x: pd.Series, title: str):
    x = x.dropna()
    if len(x) < 10:
        return
    sm.qqplot(x, line="s")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# ---------- Función principal ----------
def test_normalidad(x: pd.Series, label: str, prefix_png: str):
    jb_stat, jb_p, skew, kurt = jarque_bera_test(x)
    ks_stat, ks_p = ks_test_standardized(x)
    sw_stat, sw_p = shapiro_safe(x)

    qqplot(x, f"QQ-plot: {label}")

    return {
        "Serie": label,
        "N": len(x.dropna()),
        "JB_stat": jb_stat, "JB_p": jb_p,
        "Skew": skew, "Kurt_excess": kurt,
        "KS_stat": ks_stat, "KS_p": ks_p,
        "Shapiro_stat": sw_stat, "Shapiro_p": sw_p
    }

# ---------- Pipeline general ----------
def normalidad_precios(px: pd.Series, name_prefix: str):
    prices = px.dropna()
    log_prices = np.log(prices)
    returns = prices.pct_change().dropna()

    results = []
    results.append(test_normalidad(prices, f"{name_prefix} - Precios", f"{name_prefix}_prices"))
    results.append(test_normalidad(log_prices, f"{name_prefix} - Log(Precios)", f"{name_prefix}_logprices"))
    results.append(test_normalidad(returns, f"{name_prefix} - Rendimientos", f"{name_prefix}_returns"))
    return results

all_results = []

for ticker in ['005930.KS', 'COST', 'LLY', 'MCD', 'V']:
    series = d_port[ticker]
    all_results.extend(normalidad_precios(series, ticker))

# Portafolio
weights = np.array([0.4071, 0.1737, 0.3678, 0.0181, 0.034])
ret_pf = d_port.pct_change().dropna()
port_ret = (ret_pf * weights).sum(axis=1)
port_index = 100 * (1 + port_ret).cumprod()

all_results.extend(normalidad_precios(port_index, "PORTFOLIO"))

# Convertir resultados a DataFrame
df_normalidad = pd.DataFrame(all_results)

df_normalidad.round(4)
```

    Modelado teórico usando las ecuaciones diferenciales básicas: la ecuación dSt (modelo de movimiento geométrico browniano) y la ecuación para ln St. Explicar supuestos y limitaciones.

```python
from scipy.stats import norm

# --- Parámetros ---
TRADING_DAYS = 252
ALPHA = 0.05
Z = norm.ppf(1 - ALPHA/2)

TICKERS = ['005930.KS', 'COST', 'LLY', 'MCD', 'V']
WEIGHTS = np.array([0.4071, 0.1737, 0.3678, 0.0181, 0.034], dtype=float)
WEIGHTS = WEIGHTS / WEIGHTS.sum()

# --- Datos & log-rendimientos del portafolio ---
def get_portfolio_log_returns(tickers=TICKERS, weights=WEIGHTS, start="2015-01-01", end=None):
    px = yf.download(tickers, start=start, end=end, auto_adjust=False, progress=False)['Adj Close']
    if isinstance(px, pd.Series):
        px = px.to_frame()
    lpx  = np.log(px)
    lret = lpx.diff().dropna(how="any")                  
    port_lret = pd.Series(lret.values @ weights, index=lret.index, name="PortLogRet")
    port_index = 100.0 * np.exp(port_lret.cumsum())      # índice base=100
    return port_lret, port_index

# --- Cono de pronóstico desde log-retornos (GBM) ---
def forecast_cone_from_logrets(lret_series, S_T, H=60, z=Z):
    mu_d  = lret_series.mean()
    sig_d = lret_series.std(ddof=1)
    h = np.arange(1, H+1, dtype=float)
    mean_Rh = mu_d * h
    lo_Rh   = mean_Rh - z * sig_d * np.sqrt(h)
    hi_Rh   = mean_Rh + z * sig_d * np.sqrt(h)
    center_price = S_T * np.exp(mean_Rh)
    low_price    = S_T * np.exp(lo_Rh)
    high_price   = S_T * np.exp(hi_Rh)
    return {
        "h": h,
        "mu_d": mu_d,
        "sig_d": sig_d,
        "center_price": center_price,
        "low_price": low_price,
        "high_price": high_price
    }

# --- Graficar historial + cono (solo Figura 1) ---
def plot_cone_with_history(port_index, cone, title_prefix="Portafolio", H_freq="B", hist_window=None):
    if hist_window is not None and hist_window < len(port_index):
        hist = port_index.iloc[-hist_window:].copy()
    else:
        hist = port_index.copy()

    last_date = hist.index[-1]
    future_idx = pd.bdate_range(start=last_date, periods=int(cone["h"][-1])+1, freq=H_freq)[1:]

    # ----- Figura 1: Precio/índice con cono -----
    plt.figure(figsize=(11,6))
    plt.plot(hist.index, hist.values, label="Histórico (índice base=100)", linewidth=1.8, color="dodgerblue")
    plt.plot(future_idx, cone["center_price"], label="Trayectoria esperada", linewidth=2, color="deeppink")
    plt.fill_between(future_idx, cone["low_price"], cone["high_price"], alpha=0.20, label="Banda 95%", color="palegreen")
    plt.title(f"{title_prefix}: Historial + Cono de precios (GBM)")
    plt.ylabel("Índice / Precio")
    plt.legend()
    plt.tight_layout()
    plt.show()

# ================== Ejecutar ==================
if __name__ == "__main__":
    port_lret, port_index = get_portfolio_log_returns(start="2015-01-01")
    S_T = float(port_index.iloc[-1])
    H = 60  # horizonte de 60 días hábiles
    cone = forecast_cone_from_logrets(port_lret, S_T=S_T, H=H, z=Z)

    # Mostrar último año de historial
    plot_cone_with_history(port_index, cone, title_prefix="Portafolio (pesos dados)", hist_window=252)
```
